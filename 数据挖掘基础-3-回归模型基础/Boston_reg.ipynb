```python
import pandas as pd
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
#数据加载
x_df=pd.DataFrame(datasets.load_boston()['data'],columns=datasets.load_boston()['feature_names']) #X
y_df=pd.DataFrame(datasets.load_boston()['target'],columns=['y']) #Y
df=x_df.join(y_df)
```

```python
#画出RM,DIS,PTRATIO,LSTAT与y的散点图
plt.figure(figsize=(12,8))
plt.subplot(221)
plt.scatter(df['RM'], df['y'], color='DarkBlue')
plt.xlabel('RM')
plt.ylabel('House Price')
plt.subplot(222)
plt.scatter(df['DIS'], df['y'],color='Brown')
plt.xlabel('DIS')
plt.ylabel('House Price')
plt.subplot(223)
plt.scatter(df['PTRATIO'], df['y'], color='Yellow')
plt.xlabel('PTRATIO')
plt.ylabel('House Price')
plt.subplot(224)
plt.scatter(df['LSTAT'], df['y'], color='Red') 
plt.xlabel('LSTAT')
plt.ylabel('House Price')
plt.show()
```

由上图：
RM(平均每栋住宅数)与price(房价)存在一定的正向相关关系；
LSTAT(地区中有多少房东属于低收入人群)与price(房价)存在一定的负向相关关系；
DIS(距离5个波士顿的就业中心的加权距离)与price(房价)存在不确定的正向相关关系；
PTRATIO(城镇中教师学生比例)与price(房价)不存在明显的相关关系；

```python
#线性回归
reg = linear_model.LinearRegression()
model=reg.fit(x_df[['RM','DIS','PTRATIO','LSTAT']],y_df)
#X的系数
model.coef_
```

由上： RM(平均每栋住宅数)每增加一个单位，price(房价)平均增加约4个单位。

```python
#对某新小区,其RM=8,DIS=2,PTRATIO=12,LSTAT=22,预测该小区房价
model.predict([[8,2,12,22]])
```

由上，该小区房价中位数约为30.83
