```python
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#数据加载
df=pd.DataFrame(datasets.load_iris()['data'],columns=datasets.load_iris()['feature_names']) 
```

```python
#k-means模型训练
model = KMeans(n_clusters=3)
model.fit(df)
```

```python
#样本点聚类结果输出
predictions = model.predict(df)
```

```python
#以花萼长度(sepal length),花瓣长度(petal length)为x,y轴,可视化聚类结果
sepal_length = df.iloc[:,0]
petal_length = df.iloc[:,2]
#plt.figure(figsize=(5, 4), dpi=120)
plt.scatter(sepal_length, petal_length, c=predictions)
plt.xlabel('Sepal Length(cm)')
plt.ylabel('Petal Length(cm)')
plt.show()
```
