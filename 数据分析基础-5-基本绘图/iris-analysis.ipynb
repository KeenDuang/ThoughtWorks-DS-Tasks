```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = pd.read_csv('iris.csv')
```

```python
x1=iris['PetalLength']
y1=iris['PetalWidth']

plt.figure(figsize=(11, 8), dpi=120)
plt.scatter(x1[:50], y1[:50], color='DarkBlue', label='Iris-setosa') 
plt.scatter(x1[100:], y1[100:],color='Brown', label='Iris-Virginica') 
plt.scatter(x1[50:100], y1[50:100], color='Yellow', label='Iris-versicolor') 
plt.legend(loc=2) #左上角
plt.xlabel('SepalLength')
plt.ylabel('SepalWidth')
plt.show()
```

```python
x=iris['SepalLength']
y=iris['SepalWidth']

plt.figure(figsize=(11, 8), dpi=120)
plt.scatter(x[:50], y[:50], color='DarkBlue', label='Iris-setosa') #前50个样本
plt.scatter(x[100:], y[100:],color='Brown', label='Iris-Virginica') #后50个样本
plt.scatter(x[50:100], y[50:100], color='Yellow', label='Iris-versicolor') #中间50个
plt.legend() #右上角
plt.xlabel('SepalLength')
plt.ylabel('SepalWidth')
plt.show()
```
