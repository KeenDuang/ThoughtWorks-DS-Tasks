```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
#数据加载
X,y = datasets.load_iris(return_X_y=True) #X与y
target_names = datasets.load_iris().target_names #y的值列表:0:setosa,1:versicolor,2:virginica
feature_names = datasets.load_iris().feature_names #特征X的名称列表
```

```python
#决策树模型训练
tree_clf = DecisionTreeClassifier(max_depth=3,random_state=2020)
tree_clf.fit(X,y)
```

```python
#预测sepal length=6,sepal width=1,petal length=3,petal width=1最可能是什么类型的花
pre = tree_clf.predict(np.array([[6,1,3,1]]))
target_names[pre[0]]
```

```python
#决策树可视化
import graphviz
import pydotplus
from IPython.display import Image
import os
os.environ["PATH"] += os.pathsep + 'C:/Program Files (x86)/Graphviz2.38/bin/'
dot_data = tree.export_graphviz(tree_clf,out_file=None,feature_names=feature_names,class_names=target_names
                                ,filled=True,rounded=True,special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data)
img = Image(graph.create_png())
img
```

由上图可知，当petal width（花瓣宽度）小于0.8cm时，其最可能是setosa
