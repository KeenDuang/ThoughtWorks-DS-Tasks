```python
import pandas as pd
```

```python
tag = pd.read_csv('./ml-1m/tags.csv')
rating = pd.read_csv('./ml-1m/ratings.csv')
movie = pd.read_csv('./ml-1m/movies.csv')
```

```python
# 电影平均分加工
rating_ = pd.DataFrame(rates.groupby('movieId').mean()['rating'])
```

```python
# 电影tag加工
tag_ = pd.DataFrame(tag.groupby('movieId')['tag'].unique())
```

```python
res = pd.merge(movie,rating_,how='left',left_on='movieId',right_index=True)
res = pd.merge(res,tag_,how='left',left_on='movieId',right_index=True)
```
